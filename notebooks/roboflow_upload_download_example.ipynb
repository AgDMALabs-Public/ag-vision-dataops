# Get a list of files that you want to annotate.

# Create an annotation batch.
* This function will copy the images to be uploaded to the annotation folder generate new metadata files and rename them with a new UUID.

```python
aio.create_annotation_batch(img_list=file_listsss,
                            project_path=project_path,
                            annotation_type=annotation_type,
                            task_name=task_name,
                            batch_name=batch_name)
```

# Upoload the batch to Roboflow.
* This will look for the images in the batch and upload them to Roboflow.
* NOTE: Roboflow looks for duplicate images and not by img name but by some other method. Therefore at times there will be discrepencies between the number you uploaded and the number that made it to roboflow even if it ran successfully.
* This is nice tho, in the event it failed you can rerun and it should dedupe for you.


```python
# Roboflow will check for duplicate images and skip them on upload. That is why we have some missing images in RF.
rio.upload_image_batch_to_roboflow(rf_project=project,
                                        project_path=project_path,
                                        annotation_type=annotation_type,
                                        task_name=task_name,
                                        batch_name=batch_name,
                                        split='train',
                                        tmp_copy = True)
```
