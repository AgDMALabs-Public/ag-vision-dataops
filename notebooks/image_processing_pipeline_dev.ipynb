```python
def validate_column_values(df, value_list, column_name):
    for val in df[column_name].unique():
        if val not in value_list:
            raise ValueError(f"Value {val} not found in the standard list {value_list}")
    print(f"All values in column {column_name} are valid")
```

```python
# check the values in the df
validate_column_values(df = img_df,
                       value_list = cst.YEAR_LIST,
                       column_name = 'year')

validate_column_values(df = img_df,
                       value_list = cst.CROP_LIST,
                       column_name = 'crop')

validate_column_values(df = img_df,
                       value_list = cst.TIME_OF_YEAR_LIST,
                       column_name = 'time_of_year')

validate_column_values(df = img_df,
                       value_list = cst.COUNTRY_CODES.keys(),
                       column_name = 'country')
```
