```python
import geopandas as gpd
import pandas as pd
```

```python
file_name = '/Users/danielwilliams/Documents/Field Data/RTK_RHBV_2023A.csv'
out_name = '/Users/danielwilliams/Documents/Field Data/gcp_details.geojson'
```

```python
df = pd.read_csv(file_name)
```

```python
df.head()
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.x, df.y), crs="EPSG:4326")
```

```python
gdf.head()
```

```python
gdf.to_file(out_name, driver="GeoJSON")
```
